```python
import numpy as np   #math
import pandas as pd   
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
# movies
# movies.shape
movies.head(10)
```

```python
## merge both dataset
movies=movies.merge(credits,on="title")
movies
```

```python
movies.info()
```

```python

movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.info()
```

```python
##checking for null values

movies.isnull().sum()
print(movies.shape)
#as there is only 3 null values so deleting 3 rows
movies.dropna(inplace=True)
movies
```

```python
## check for duplicate data
movies.duplicated().sum()
```

```python
## bring useful format 
## take genres column
import ast
movies.iloc[0].genres
## we need only action,comedy,adventure....

def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
## ast is used to convert string list to original list in python
## ast.literal_eval(String list)

## to apply Convert function to genres 
movies['genres']=movies['genres'].apply(convert)

##use same function for keywords column
movies['keywords']=movies['keywords'].apply(convert)

movies.head(1)
```

```python
## make a function to fetch top three cast("heros")
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i["name"])
            counter+=1
        else:
            break
    return L

## apply function to cast
movies['cast']=movies['cast'].apply(convert3)
```

```python
## fetching director from crew
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            L.append(i['name'])
            break
    return L

## apply to crew
movies["crew"]=movies['crew'].apply(fetch_director)
```

```python
## to convert overview column(paragraph to list) so we can concate this column with other columns

movies['overview']= movies['overview'].apply( lambda x : x.split())
```

```python
##remove space from data
##like Sam Worthington ----to---- SamWorthington 
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ",'') for i in x])
movies['overview'] =movies['overview'].apply(lambda x: [i.replace(" ",'') for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ",'') for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ",'') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ",'') for i in x])
```

```python
movies['genres']
```

```python
## concate all columns except movie_id and title in tags
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# create a new dataframe and fetch only movie_id, title, tags

new_df=movies[['movie_id',"title","tags"]]
new_df.iloc[0].tags
```

```python
## convert list to string in tags
new_df["tags"]=new_df['tags'].apply(lambda x : " ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

# Vectorisation

```python
## we have to convert tags into vector so that it is easy to show similar movies by vector (2D array)
## we are using 'bags of words method here'

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vector=cv.fit_transform(new_df["tags"]).toarray()
cv.get_feature_names_out()
```

```python
vector
vector[1]
```

```python
## to see 5000 words whoch is generated in tags column
cv.get_feature_names_out()
```

```python
## we are getting corpus(5000 words) like love..loved..loving but we need only love 
## so we have to stem it using natural prossesing library called nltk
!pip install nltk
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()

## make a function for steming
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new_df["tags"]=new_df["tags"].apply(stem)
```

```python
cv.get_feature_names_out()
```

## similar movies fetching

```python
## we are calculating cosine angle between the vectors for checking the angular distance between two vectors(movies)
## we are not using euclidian measurement technique here that is not good for higher dimensional array

from sklearn.metrics.pairwise import cosine_similarity
## 1--similar
## 0-- not similar

similarity=cosine_similarity(vector)
similarity[0]
```

# make function to fetch top 5 movies

```python
#### edit 
new_df['title']=new_df['title'].apply(lambda x: x.lower())
```

```python
def recommend(movie):
    movie=movie.lower()
    movie_index=new_df[new_df["title"]==movie].index[0]
    
    ## calculating all movies similarity between 0 to 1
    distances=similarity[movie_index]
    ## making list of top 5 similar movies 
    movies_list=sorted(list(enumerate(distances)),reverse=True,key =lambda x: x[1])[1:6]
    
    ## to fetch top 5 similarity of movies
    for i in movies_list:
        ##print(i[0])
        print((new_df.iloc[i[0]].title).capitalize())
        
recommend('AVatar')
        
        
```

```python
recommend("the Dark Knight Rises")
```
